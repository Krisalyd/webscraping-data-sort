```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import os

url = 'https://www.ncei.noaa.gov/data/local-climatological-data/access/2021/'



if not os.path.exists('data'):
    os.makedirs('data')
    print('Directory "data" created')

print(f'Waiting response from GET request {url}')
html = requests.get(url).content
soup = BeautifulSoup(html, 'html.parser')
csv_file = soup.find('td', string='2022-02-07 14:03  ').find_previous_sibling('td').text
download_url = requests.get(url+csv_file).content
print(f'Downloading file {csv_file}')
open(f'data/{csv_file}', 'wb').write(download_url)
print(f'File {csv_file} downloaded.')



```

```python
data_pd = pd.read_csv(f'data/{csv_file}')
filtered_data = data_pd.sort_values(['HourlyDryBulbTemperature'], ascending=False).head(5)
filtered_data
    
```

```python
filtered_data.to_csv(f'Filter_{csv_file}', index=True)
```
